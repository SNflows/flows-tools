```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from astropy.cosmology import FlatLambdaCDM
cosmo = FlatLambdaCDM(H0=70, Om0=0.3, Tcmb0=2.725)

%config InlineBackend.figure_format = 'retina'
plt.rcParams["font.family"] = "Gentium"
plt.rcParams['mathtext.fontset'] = "cm"
```

```python
band = 'J'
params_df = pd.read_csv('sn_parameters.csv')
```

```python
z = params_df.z.values
mag = params_df[f'{band}max'].values
mag_err = params_df[f'{band}max_err'].values

fig, ax = plt.subplots(figsize=(8, 6))
    
    
# plot
ax.errorbar(z, mag, yerr=mag_err, 
            fmt='o', color='gray', alpha=0.4)
   
# ax.axhline(0, color='k', ls='--', zorder=20)
ax.set_xlabel('$z$', fontsize=26)
ax.set_ylabel(r'$%s_{\rm max}$'%band, fontsize=24)
ax.set_title('Hubble Diagram', fontsize=24)
ax.tick_params(labelsize=20)

fig.tight_layout()
#plt.savefig(f'paper_plots/{param}_residuals.jpg')
plt.show()
```
